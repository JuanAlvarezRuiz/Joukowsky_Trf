```python
import numpy as np
import matplotlib.pyplot as plt

# Joukowski Transformation function (takes coordinates a,b and returns new coordinates x,y via the complex trf z=\zeta+1/\zeta)
def joukowski_transformation(zeta):
    a=zeta[0,:]
    b=zeta[1,:]

    return np.array([a*(1+1/(a**2+b**2)),b*(1-1/(a**2+b**2))])

#Define the \zeta coordinates that we want, it will be circle centered in x0,y0 that goes through (1,0)
x0=-0.5
y0=0.5
r=np.sqrt((1-x0)**2+y0**2)
# Generate points on the circle that goes through (1,0) and has center (x0,y0) 
theta = np.linspace(0, 2 * np.pi, 500)
circle_points = np.array([x0+r*np.cos(theta),y0+r*np.sin(theta)])  
# Apply the Joukowski transformation
transformed_points = joukowski_transformation(circle_points)

# Plot the original circle and the transformed airfoil shape
plt.figure(figsize=(8, 6))

# Plot the original circle
plt.plot(circle_points[0,:], circle_points[1,:], label=r"Original Circle (in $\zeta$ plane)", color="blue", linestyle="--")

# Plot the transformed shape (airfoil)
plt.plot(transformed_points[0,:], transformed_points[1,:], label="Transformed Airfoil", color="red")

plt.gca().set_aspect('equal', adjustable='box')
plt.title("Joukowski Transformation: Circle to Airfoil")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
